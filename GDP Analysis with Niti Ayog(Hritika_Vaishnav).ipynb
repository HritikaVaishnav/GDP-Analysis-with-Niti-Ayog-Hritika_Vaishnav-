```python
# Filtering out the warnings

import warnings

warnings.filterwarnings('ignore')

# Import the required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing data from CSV file into pandas dataframe(gdp_current_price)

gdp_current_price=pd.read_csv("Gross Domestic Product (GDP) at Current Price.csv")

gdp_current_price
```

```python
# Pivot_table is use for trasform and reshape the data(And also organize the data)

pd.pivot_table(gdp_current_price,index=['Items Description','Duration'])
```

```python
#Drop(Removing) unnecessary Indexes

gdp_current_price=gdp_current_price.drop(gdp_current_price.index[[3,9]]) # Beacuse index 3, 9 has many NaN values

gdp_current_price
```

```python
# To get a concise summary of the dataframe.

gdp_current_price.info()
```

```python
# The DataFrame columns attribute provides the label values for columns.

column_attributes=list(gdp_current_price.columns)
```

```python
del(column_attributes[0]) # deleting Items Description 
del(column_attributes[0]) # deleting Duration
```

```python
states=column_attributes
states
```

```python
# To get Average growth rate for each state use mean() function.
# Mean of the row (% Growth over previous year) for duration 2013-14, 2014-15 and 2015-16.

state_mean={}
for column in gdp_current_price:
    if column in states:
        mean=gdp_current_price[gdp_current_price["Items Description"]=="(% Growth over previous year)"][column].mean()
        state_mean[column]=mean
state_mean["Items Description"]="Average growth rate"
gdp_current_price=gdp_current_price.append(state_mean,ignore_index=True)
gdp_current_price
```

```python
# To select a value that belongs to a particular row or column from a set of values of a data frame or dataset.

gdp_current_price=gdp_current_price.iloc[[2,7,8],:]
gdp_current_price
```

```python
#For better understanding making rows as columns and vice versa using T.
gdp=gdp_current_price.T
gdp
```

```python
gdp.columns=gdp.iloc[0]
```

```python
gdp=gdp[1:]
```

```python
gdp=gdp[1:]
gdp
```

```python
# Giving the path of all the csv files

path = ("E:\\Hritika Vaishnav Tak\\lockdown Data\\Decodr\\Case Study\\Case Study 3 - GDP Analysis with Niti Ayog\\Dataset\\GSVA\\")
listOfFiles=os.listdir(path)
```

```python
len(listOfFiles)
```

```python
# Reading all the csv files using ListOfFiles functionality from a directory for further analysis

for filename in listOfFiles:
    for statename in states:
        if statename in filename:
            filepath=path+filename
            data=pd.read_csv(filepath,encoding="ISO-8859-1")
            data=data[['Item','2014-15']]
            data=data.T
            data.columns=data.iloc[0]
            data=data[1:]
            data["state"]=statename
            gsva=pd.concat([data,gsva])
```
